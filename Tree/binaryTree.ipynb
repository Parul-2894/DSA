Binary Tree: A regular tree but each node can have at max only 2 nodes.

Binary search tree is a a special type f binary tree where nodes on the left side are smaller then the root node and the nodes on the right side are greater than the root node.

Elements are always unique.

## Traversal Technique 
<ol>
   <li> Bredth first search</li>
    <li> Depth first search 
        <ol>
        <li>inoder traversal - left root right</li>
        <li>preoder traversal  root left right</li>
        <li>post order traversal - left right root</li>
        </ol>
    </li>
</ol>

```python
class BinarySearchTreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None

    def add_child(self, data):
        if data == self.data:
            return
        
        if data < self.data:
            #add on left side 
            if self.left:
                self.left.add_child(data)
            else:
                self.left = BinarySearchTreeNode(data)
                return

        else:
            #add on right side
            if self.right:
                self.right.add_child(data)
            else:
                self.right = BinarySearchTreeNode(data)
                return
            
    def inorder_traversal(self):
        elements = []

        #traverse left
        if self.left:
            elements += self.left.inorder_traversal()
        
        #visit base
        elements.append(self.data)

        #traverse right
        if self.right:
            elements += self.right.inorder_traversal()
        
        return elements
    
    def search(self,data):
        if self.data == data:
            return "Found", self.data
        elif data<self.data:
            if self.left:
                return self.left.search(data)
            else:
                return "Not Found"
        else:
            if self.right:
                return self.right.search(data)
            else:
                return "Not Found"
            
    def find_max(self):
        if self.right is None:
            return self.data
        else:
            return self.right.find_max()

    def find_min(self):
        if self.left is None:
            return self.data
        else:
            return self.left.find_min()
        
    def delete(self, val):
        print("Vals is ")
        print(val)
        print('self is ')
        print(self.data)
        if val<self.data:
            if self.left:
                self.left = self.left.delete(val)

        elif val>self.data:
            if self.right:
                self.right = self.right.delete(val)

        else:
            if self.left is None and self.right is None:
                return None
            if self.left is None:
                return self.right
            if self.right is None:
                return self.left
            
            min_val  = self.right.find_min()
            self.data = min_val
            self.right = self.right.delete(min_val)
        return self
    
    def delete_alternate_using_max_from_left(self, val):
        if val < self.data:
            self.left = self.left.delete_alternate_using_max_from_left(val)

        elif val > self.data:
            self.right = self.right.delete_alternate_using_max_from_left(val)

        else:
            if self.left is None and self.right is None:
                return None
            if self.right is None:
                return self.left
            if self.left is None:
                return self.right
            
            max_val_left = self.left.find_max()
            self.data = max_val_left
            self.left = self.left.delete_alternate_using_max_from_left(max_val_left)
        return self

    
a = [15, 16,4,3,1,17,20]
b1 = BinarySearchTreeNode(15)
b1.add_child(16)
b1.add_child(4)
b1.add_child(3)
b1.add_child(1)
b1.add_child(17)
b1.add_child(20)

print(b1.inorder_traversal())
print(b1.search(16))
b1.delete(15)
print(b1.inorder_traversal())
b1.delete(3)
print(b1.inorder_traversal())


```

## Delete a node in binary Search tree

Deleting a node in a binary search tree. 
if its a leaf node delete it simple. 
if it has only one child we can connect that one child to the parent and remove it. 
if itt has two child then we have two alternative ways; 
1. either we can take the minumum value of the right sub tree and move it to the elements place
2. or we can take the maximum value of the maximum value of the right subtree and move it to the elements place.
