Array insertion complexity is of O(n)
Python list is a dynamic array. 
if you insert an element at a certain location, the next two value are first moved to next location and then a value is inserted. 

Array store values in contiguous location.
But in linked list we can store values at different location and then just give that link to the previous node. 

Since here no contiguous memory is required. 

Insert elemetn at the beginning - O(1)
Delete elemetn at the beginning - O(1)
insert/delete Elementt at the end = O(n)
Linked ist Traversal = O(n)
Accessing Element by Value = O(n)


```python
class Node:    #represents a single nde in a linked list. it has two class members
    def __init__(self, data=None, next=None):
        self.data = data    #any value  like int, string anything
        self.next = next    #pointer to the next element 


#this is a linked list class which starts with pointing to  the header.
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert_at_beginning(self, data):
        node = Node(data, self.head)
        self.head = node

    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data, None)
        else:
            node = Node(data,None)
            itr = self.head
            while itr.next:
                itr = itr.next
            itr.next = node

    def create_linked_list_from_list(self,data_list):
        self.head = None
        for i in data_list:
            self.insert_at_end(i)
        self.print()
        
    
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count+=1
            itr = itr.next
        return count

    def print(self):
        if self.head is None:
            print("linked list is empty ")
            return
        
        itr = self.head
        llstr = ''
        while itr:
            llstr += f'{itr.data}--->'
            itr = itr.next
        print(llstr)
        

    def remove_at(self, index):

        if index<0 or index>=self.get_length():
            raise Exception("Invalid index")
        
        if index == 0:
            self.head = self.head.next
            return
        
        count  = 0
        itr = self.head
        while count<index-1:
            count += 1
            itr  = itr.next
        itr.next  = itr.next.next

    def insert_at(self,index, data):
        if index<0 or index>=self.get_length():
            raise Exception("Invalid index")
        

        if index == 0:
            node = Node(data, self.head)
            self.head = node
            return 
        
        count = 0

        itr = self.head
        while itr:
            if count== index-1:
                break
            itr = itr.next
            count+=1
        node = Node(data, itr.next)
        itr.next = node
    
    def insert_after_value(self, data_after, data_to_insert):
        itr = self.head
        value_added = False
        while itr:
            if itr.data == data_after:
                node = Node(data_to_insert, itr.next)
                itr.next = node
                value_added=True
            itr = itr.next
            
        
        if not value_added:
            return "No Data after found"
        else:
            return "Data Added"
        
    def remove_by_value(self, data):
        itr = self.head
        if itr.data==data:
            self.head = itr.next
        while itr.next:
            if itr.next.data == data:
                itr.next = itr.next.next
            itr = itr.next




node = Node(5)
ll = LinkedList()
# ll.insert_at_beginning(5)
# ll.insert_at_beginning(4)
# ll.insert_at_beginning(2)
# ll.insert_at_end(10)
# ll.insert_at_end(12)
# ll.insert_at_end(13)
ll.create_linked_list_from_list(["banana", "apple", "mango", "oranges","kiwi"])
ll.print()
# ll.remove_at(3)
ll.insert_at(4,"grapes")
ll.print()
ll.insert_after_value("banana","guava")
ll.remove_by_value("apple")
ll.print()

print(ll.get_length())
```
